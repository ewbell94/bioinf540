```python
import matplotlib.pyplot as p
from scipy.stats import pearsonr
import numpy as np
import pandas as pd
```

```python
f=open("results.list")
lengths=[]
fnums=[]
errors=[]
df = pd.DataFrame(columns = ["Name", "Length", "Fnum", "Error", "else"])
```

```python
for line in f:
    parts=line.split(',')
    lengths.append(int(parts[1]))
    fnums.append(float(parts[2]))
    nmerr = abs(float(parts[3])-float(parts[4]))/int(parts[1])
    errors.append(nmerr)
    df.loc[-1] = [parts[0], int(parts[1]), float(parts[2]), float(nmerr), float(parts[4])]
    df.index = df.index + 1
    df = df.sort_index() 
```
